```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Specify the file path
file_path = '/Users/arushi/Desktop/Market Basket Analysis/Online Retail.xlsx'

# Load the Excel file
myretaildata = pd.read_excel(file_path)
```

```python
myretaildata.head()
```

```python
myretaildata.isna().sum()
```

## Data Prepration

```python
myretaildata['Description']=myretaildata['Description'].str.strip()
myretaildata.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
myretaildata['InvoiceNo']=myretaildata['InvoiceNo'].astype('str')
myretaildata=myretaildata[~myretaildata['InvoiceNo'].str.contains('C')]
myretaildata.head()
```

```python
myretaildata['Country'].value_counts()
```

```python
myretaildata.shape
```

```python
#seperating transactions for Germany
mybasket = (myretaildata[myretaildata['Country']=="Germany"]
           .groupby(['InvoiceNo', 'Description'])['Quantity']
           .sum().unstack().reset_index().fillna(0)
           .set_index('InvoiceNo'))
```

```python
#viewing transaction basket
mybasket.head()
```

```python
# converting all positive values to 1 and everything else to zero
def my_encode_units(x):
    if x <=0:
        return 0
    if x >=1:
        return 1
my_basket_sets=mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE',inplace=True, axis=1)
```

## Training the Model

```python
#generating frequent itemsets
my_frequent_itemsets=apriori(my_basket_sets,min_support=0.07, use_colnames=True)
```

```python
#generating rules
my_rules=association_rules(my_frequent_itemsets, metric="lift",min_threshold=1)
```

```python
#viewing top 100 views
my_rules.head(100)
```
